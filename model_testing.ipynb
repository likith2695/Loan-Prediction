```python
import pandas as pd
import numpy as np
```

```python
test_df= pd.read_csv('Test Data.csv')
```

```python
import pickle

filename = 'Loan_Prediction.sav'
rf = pickle.load(open(filename, 'rb'))

mar_sin = pickle.load(open('mar_sin.sav', 'rb'))
house_own = pickle.load(open('house_own.sav', 'rb'))
car_own = pickle.load(open('car_own.sav', 'rb'))
prof = pickle.load(open('prof.sav', 'rb'))
city = pickle.load(open('city.sav', 'rb'))
state = pickle.load(open('state.sav', 'rb'))
```

```python
test_df.head()
```

```python
test=test_df.drop(columns=['ID'],axis=1)
```

```python
test.head()
```

```python
test['Married/Single']=mar_sin.transform(np.array(test['Married/Single']).reshape(-1,1))
test['House_Ownership']=house_own.transform(np.array(test['House_Ownership']).reshape(-1,1))
test['Car_Ownership']=car_own.transform(np.array(test['Car_Ownership']).reshape(-1,1))
test['Profession']=prof.transform(np.array(test['Profession']).reshape(-1,1))
test['CITY']=city.transform(np.array(test['CITY']).reshape(-1,1))
test['STATE']=state.transform(np.array(test['STATE']).reshape(-1,1))
```

```python
test.head()
```

```python
test_pred = rf.predict(test)
```

```python
test_pred
```

```python
test_df_1 = test_df[['ID']]
test_df_1['risk_flag'] = test_pred
```

```python
test_df_1.head()
```

```python
test_df_1.to_csv('prediction.csv',index=False)
```
